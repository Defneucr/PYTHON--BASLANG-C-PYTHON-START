```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
class öğrenci:
    def __init__(self,isim,soy_isim,doğum_tarihi,doğum_yeri):
        self.isim = isim
        self.soy_isim = soy_isim.lower()
        self.doğum_tarihi=doğum_tarihi.lower()
        self.doğum_yeri=doğum_yeri.lower()
        self.şifre_tablosu=pd.DataFrame(columns=['Date','Şifre'])
        self.şifre_kurtarma_sorusu=[]
        
    def tüm_şifreleri_göster(self):
        return self.şifre_tablosu
    def son_şifre_göster(self):
        return self.şifre_tablosu.loc[şifre_tablosu.shape[0]-1,:]
        
    def şifre_oluştur (self):
        if self.şifre_tablosu.shape[0]==0:
            
            input_1 =input ('Lütfen isim ve soy isminizi küçük harflerle giriniz:').lower()
            input_2= input ('Lütfen doğum yerinizi küçük harflerle giriniz:').lower()
            input_3= input ('Lütfen doğum tarihinizi gün ay ve yıl şeklinde giriniz:')

            if (input_1!=str(self.isim.lower())+' '+str(self.soy_isim.lower())) | (input_2!=self.doğum_yeri.lower()) | (input_3!=self.doğum_tarihi) :
                print('Girdiğiniz bilgiler kayıtlarımlla uyuşmuyor, lütfen tekrar deneyin')


            else:

                input_4=input('Lütfen 8 ile 12 karakterden oluşan, en az bir büyük, en az bir küçük ve en az bir özel karakter içeren,\
                isim ve soy isminizle başlayıp bitmeyen ve son bir yıl içersinde kullanmadığınız bir şifre giriniz:')
                şifre = str(input_4)
                harf_sayısı =0
                özel_karakter_sayısı=0
                # higher or equal to 8 and lower or equal to 12
                if (len(şifre) < 8) | (len(şifre)>12):

                    print("Şifreniz 8 haneden büyük veye 12 haneden küçük olmadığı için kabul edilmedi.")


                elif (şifre[:(len(self.isim))]==self.isim) | (şifre[-len(self.isim):]==self.isim):
                    print('Şifreniz isminizle başladaığı veya bittiği için, şifreniz kabul edilmedi')
                
                elif (şifre[:(len(self.soy_isim))]==self.soy_isim) | (şifre[-len(self.soy_isim):]==self.soy_isim):
                    print('Şifreniz soy isminizle başladığı veya bittiği için, şifreniz kabul edilmedi')
                
                else:
                    for i in şifre:
                        if i in ['#','%','$','@','~','*','&']:
                            özel_karakter_sayısı=özel_karakter_sayısı+1
                    if özel_karakter_sayısı<1:
                        print("Şifreniz en az bir özel karakter içermediğiden dolayı kabul edilmedi")
                    
                    else:
                        for i in şifre:
                            if i.isalpha():
                                harf_sayısı = harf_sayısı +1
                        if harf_sayısı<2:
                            
                            print("Şifreniz en az bir harf içermediğinden dolayı kabul edilmedi")
                        
                        else:
                            input_şifre_kurtarma=input('Lütfen şifrenizi unuttuğunuz zaman kullanmak üzere, bu soruyu cevaplayın.\
                            İlkokulu hangi okulda okudunuz?').lower()
                            print("Şifre kabul edildi")
                            self.şifre_tablosu.loc[self.şifre_tablosu.shape[0],'Date']=datetime.now()
                            self.şifre_tablosu.loc[self.şifre_tablosu.shape[0]-1,'Şifre']=şifre
                            self.şifre_kurtarma_sorusu.append(input_şifre_kurtarma)
        else:
            print('İlk şifre oluşumu yapılmıştır, eğer şifrenizi unutursanız yeni bir şifre oluşturmak için\
            veya şifremi unuttum linkine basınız.Eğer mevcut şifrenizi değiştirmak istiyorsanız kütfen şifre değiştir linkine basınız.')

    def şifremi_unuttum(self):
        input_şifre_kurtarma1=input('Lütfen okuduğunuz ilkokulun ismini yazınız?')

        if input_şifre_kurtarma1!=self.şifre_kurtarma_sorusu[0]:
            print('Cevabınız kayıtlarımızla uyuşmuyor, lütfen kontrol edip tekrar deneyiniz')
        else:
            
                input_a=input('Lütfen 8 ile 12 karakterden oluşan, en az bir büyük, en az bir küçük ve en az bir özel karakter içeren, \
                isim ve soy isminizle başlayıp bitmeyenve son bir yıl içersinde kullanmadığınız bir şifre giriniz:')
                şifre = str(input_a)
                harf_sayısı =0
                özel_karakter_sayısı=0
                # higher or equal to 8 and lower or equal to 12
                if (len(şifre) < 8) | (len(şifre)>12):

                    print("Şifreniz 8 ile 12 haneden oluşmadığı için kabul edilmedi.")


                elif (şifre[:(len(self.isim))]==self.isim.lower()) | (şifre[-len(self.isim):]==self.isim.lower()):
                    print('Şifreniz isminizle başladığı veya bittiği için, şifreniz kabul edilmedi')
                elif (şifre[:(len(self.isim))]==self.soy_isim.lower()) | (şifre[-len(self.isim):]==self.soy_isim.lower()):
                    print('Şifreniz isminizle başladığı veya bittiği için, şifreniz kabul edilmedi')
                else:
                    for i in şifre:
                        if i in ['#','%','$','@','~','*','&']:
                            özel_karakter_sayısı=özel_karakter_sayısı+1
                    if özel_karakter_sayısı<1:
                        print("Şifreniz en az bir özel karakter içermediğiden dolayı kabul edilmedi")
                    else:
                        for i in şifre:
                            if i.isalpha():
                                harf_sayısı = harf_sayısı +1
                        if harf_sayısı<2:
                            print("Şifreniz en az bir harf içermediğinden dolayı kabul edilmedi")
                        else:
                            
                            print("Şifre kabul edildi")
                            self.şifre_tablosu.loc[self.şifre_tablosu.shape[0],'Date']=datetime.now()
                            self.şifre_tablosu.loc[self.şifre_tablosu.shape[0]-1,'Şifre']=şifre

            
    
    def şifre_değiştir (self):
        input_5=input('Lütfen şifrenizi giriniz:')
        
        if input_5 not in self.şifre_tablosu.loc[self.şifre_tablosu.shape[0]-1,'Şifre']:
            print('Girdiğiniz şifre bizde kayıtlı olan şifre ile uyuşmamaktadır')
        else:       

            input_6=input('Lütfen 8 ile 12 karakterden oluşan, en az bir büyük, en az bir küçük ve en az bir özel karakter içeren,\
            isim ve soy isminizle başlayıp bitmeyenve son bir yıl içersinde kullanmadığınız bir şifre giriniz:')
            yeni_şifre = str(input_6)
            harf_sayısı =0
            özel_karakter_sayısı=0
            # higher or equal to 8 and lower or equal to 12
            if (len(yeni_şifre) < 8) | (len(yeni_şifre)>12):

                print("Şifreniz 8 ile 12 haneden oluşmadığı için kabul edilmedi.")


            elif (yeni_şifre[:(len(self.isim))]==self.isim.lower()) | (yeni_şifre[-len(self.isim):]==self.isim.lower()):
                print('Şifreniz isminizle başladaığı veya bittiği için, şifreniz kabul edilmedi')
            elif (yeni_şifre[:(len(self.isim))]==self.soy_isim.lower()) | (yeni_şifre[-len(self.isim):]==self.soy_isim.lower()):
                print('Şifreniz isminizle başladığı veya bittiği için, şifreniz kabul edilmedi')
                
            elif yeni_şifre in self.şifre_tablosu.set_index('Date').loc[str(datetime.now().year)][['Şifre']].values[self.şifre_tablosu.shape[0]-3:]:
                print('Girdiğiniz şifre son bir yıl içinde kullandığınız son üç şifreden biri olmamalıdır')
            else:
                for i in yeni_şifre:
                    if i in ['#','%','$','@','~','*','&']:
                        özel_karakter_sayısı=özel_karakter_sayısı+1
                if özel_karakter_sayısı<1:
                    print("Şifreniz en az bir özel karakter içermdiğiden dolayı kabul edilmedi")
                else:
                    for i in yeni_şifre:
                        if i.isalpha():
                            harf_sayısı = harf_sayısı +1
                    if harf_sayısı<2:
                        print("Şifreniz en az bir harf içermediğinden dolayı kabul edilmedi")
                    else:
                        print("Şifre kabul edildi")
                        self.şifre_tablosu.loc[self.şifre_tablosu.shape[0],'Date']=datetime.now()
                        self.şifre_tablosu.loc[self.şifre_tablosu.shape[0]-1,'Şifre']=yeni_şifre



```
